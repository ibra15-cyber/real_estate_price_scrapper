```python
import requests
from bs4 import BeautifulSoup

request=requests.get("https://www.century21.com/real-estate/london-ky/LCKYLONDON/") #page 1

data=request.content
soup=BeautifulSoup(data, "html.parser")
# soup.prettify
all=soup.find_all("div")
# print(all)
```

```python
all=soup.find_all("div", {"id":"results-container"}) #all search

#each card 
#getting the first card
card=soup.find_all("div", {"class":"property-card-primary-info"})
# card
```

```python
l=[]

for item in card:
    d={} #when dictionary is global the on each iteration, the dictionary contents gets replaced so you end up with only one data

    d["Addresses"]=(item.find_all("div", {"class": "property-address"})[0].text.replace("\n", "").replace(" ",''))
    
    d["Beds"]=(item.find("div", {"class":"property-beds"}).find("strong").text.replace("\n", "").replace(" ",''))
    #the above line can be used to scrape the next line and vice versa
    #the below line is saying grab the index of the next strong tag
    d["Baths"]=(item("strong", {})[1].text.replace("\n", "").replace(" ",''))
    try:
# print(item("strong", {})[2].text.replace("\n", "").replace(" ",'')) #get the 3rd strong tag
        d["Property Size"]=(item.find_all("div", {"class": "property-sqft"})[0].find("strong").text.replace("\n", "").replace(" ",''))
    except:
        d["Property Size"]=None
        
    #     print(item.find_all("a", {"class":"listing-price"})[0].text.replace("\n", "").replace(" ",''))
    d["Prices"]=item.find("a", {"class":"listing-price"}).text.replace("\n", "").replace(" ",'')
    print("\n")
    l.append(d)
    
#find_all will take an index
#find does not need an index
```

```python
print(l)
len(l)
```

```python
import pandas
df=pandas.DataFrame(l) #passing our list of dictionary or objects to a pandas dataframe 
df
```

```python
df.to_csv("data.csv") #saving it as a csv file
```
